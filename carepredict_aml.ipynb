```python
from sklearn.linear_model import LogisticRegression

print("===== Disease Risk Prediction =====")

# Features:
# [Age, Gender, Weight, Glucose, BP, Temperature,
#  StomachPain, HeartRate, SleepHours, Activity]

X = [
[18,0,55,90,120,98,0,72,8,1],
[45,1,80,170,150,98,0,88,6,0],
[50,1,85,165,165,98,0,92,5,0],
[22,0,60,95,120,102,0,95,7,1],
[25,1,58,100,118,103,0,96,6,1],
[35,0,75,115,130,98,1,78,6,0],
[28,1,70,100,125,98,1,75,7,0],
[48,1,88,180,155,99,0,90,5,0]
]

# Labels
# 0 Healthy
# 1 Diabetes
# 2 Hypertension
# 3 Viral Fever
# 4 Gastritis

y = [0,1,2,3,3,4,4,1]

# Train
model = LogisticRegression(max_iter=1000)
model.fit(X, y)

# Inputs
age = int(input("Enter Age: "))

print("Gender → Female=0  Male=1")
gender = int(input("Enter Gender: "))

weight = float(input("Enter Weight (kg): "))
glucose = float(input("Enter Glucose Level: "))
bp = float(input("Enter Blood Pressure: "))
temp = float(input("Enter Temperature (F): "))
stomach = int(input("Stomach Pain? (1=Yes,0=No): "))
heart = int(input("Enter Heart Rate: "))
sleep = float(input("Sleep Hours per Day: "))
activity = float(input("Physical Activity Hours per Day: "))

user = [[
    age,
    gender,
    weight,
    glucose,
    bp,
    temp,
    stomach,
    heart,
    sleep,
    activity
]]

# Predict
prediction = model.predict(user)[0]
risk = round(max(model.predict_proba(user)[0]) * 100, 2)

disease = {
0: "Healthy",
1: "Diabetes",
2: "Hypertension",
3: "Viral Fever",
4: "Gastritis"
}

# Risk category
if risk < 40:
    category = "LOW"
elif risk < 75:
    category = "MODERATE"
else:
    category = "HIGH"

# Suggestions (4 sentences)
suggestion = {

"Healthy":
"""
Your values appear closer to a healthy range. Continue eating balanced meals and remain physically active. Maintain proper sleep and hydration every day. Regular health checkups can help maintain long-term wellness.
""",

"Diabetes":
"""
The prediction suggests possible diabetes-related risk. Reduce sugar intake and prefer fiber-rich foods in daily meals. Monitor glucose levels regularly and maintain body weight. Consult a healthcare professional for proper evaluation.
""",

"Hypertension":
"""
The result indicates possible high blood pressure risk. Reduce salt intake and manage stress levels consistently. Include regular exercise and healthy sleep habits. Seek medical advice if blood pressure remains elevated.
""",

"Viral Fever":
"""
The result suggests symptoms similar to viral fever. Get enough rest and drink sufficient fluids throughout the day. Eat light and nutritious food while monitoring temperature. Contact a doctor if symptoms continue or worsen.
""",

"Gastritis":
"""
The prediction suggests possible gastric irritation. Avoid spicy and oily foods and eat smaller meals. Reduce caffeine intake and stay hydrated during the day. Consult a doctor if stomach discomfort continues.
"""
}

result = disease[prediction]

print("\n========== RESULT ==========")
print("Predicted Disease :", result)
print("Risk Percentage   :", risk, "%")
print("Risk Category     :", category)
print("\nSuggestion:")
print(suggestion[result])

print("============================")
```
